```python
%matplotlib inline
import matplotlib.pyplot as plt
import utils
import models
```

### Util functions

```python
def plot_images(batch, nb_imgs):
    plt.figure(figsize=(20, 4))
    for i in range(nb_imgs):
        ax = plt.subplot(2, nb_imgs, i + 1)
        plt.imshow(batch[i].squeeze())
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()

def plot_results(mean_loss, batch_data, nb_imgs=10):
    (x, y, y_hat) = batch_data
    print(f'Mean loss: {round(mean_loss, 4)}')
    print('Input')
    plot_images(x, nb_imgs)
    print('Target')
    plot_images(y, nb_imgs)
    print('Generated')
    plot_images(y_hat, nb_imgs)
```

# Cifar

```python
(cifar_train, cifar_eval, cifar_test) = utils.load_dataset('cifar')
```

```python
dense_autoencoder = models.DenseAutoencoder((32, 32, 3))
dense_autoencoder.load_weights('trained_models/dense_autoencoder_d_cifar_b_128_e_20')
plot_results(*dense_autoencoder.evaluate(cifar_test))
```

```python
conv_autoencoder = models.ConvAutoencoder((32, 32, 3))
conv_autoencoder.load_weights('trained_models/conv_autoencoder_d_cifar_b_128_e_20')
plot_results(*conv_autoencoder.evaluate(cifar_test))
```

```python
u_net = models.UNet((32, 32, 3))
u_net.load_weights('trained_models/u_net_d_cifar_b_128_e_20')
plot_results(*u_net.evaluate(cifar_test))
```
